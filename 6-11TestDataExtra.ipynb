```python
import re
#处理另外几个不太好处理的特征
def get_list_infos_first_type(infos):
    type_times = {}
    
    for info in infos:
        info_list = str(info).split("|")
        
        if info_list[0] in type_times.keys():
            type_times[info_list[0]] = type_times[info_list[0]] + 1
        else:
            type_times[info_list[0]] = 1
                    
    return type_times



# 获取最大类别数, 以及[类别,出现次数]的字典表
def get_list_infos(infos):
    max_group = 0
    type_times = {}
    
    for info in infos:
        info_list = str(info).split("|")
        max_group = max(max_group, len(info_list))
        
        for tp in info_list:
            
            if tp in type_times.keys():
                type_times[tp] = type_times[tp] + 1
            else:
                type_times[tp] = 1
                    
    return max_group, type_times

# 讲dict中的次数信息转成概率信息
def get_dict_info_to_probability(dict_info):
    
    type_probability = {}
    
    max_times = 0
    
    for value in dict_info.values():
        max_times += value
    
    print ("总次数:", max_times)
    
    for key in dict_info:
        if max_times == 0:
            type_probability[key] = 0
        else:
            type_probability[key] = dict_info[key]/max_times
                    
    return type_probability
```

```python
#先处理歌曲类别
genreIds = df['genre_ids'].unique()
max_genre_group, genre_times = get_list_infos(genreIds)
genre_probability = get_dict_info_to_probability(genre_times)
```

```python
def get_names(name, max_artist_name_group):
    names = []
    
    for i in range(0, max_artist_name_group):
        names.append(name+'_'+str(i))
    
    return names
```

```python
from sklearn.feature_extraction import FeatureHasher
```

```python
cols = ['artist_name', 'composer', 'lyricist']
for col in cols:
    col_names = df[col].unique()
    max_col_group, col_times = get_list_infos(col_names)

    df = pd.concat(objs = [df, df[col].str.split('|', expand=True)], axis=1)

    column_names = get_names(col, max_col_group)
    df.rename(columns = dict(zip(range(0, max_col_group), column_names)), 
              inplace = True)


    df[column_names] = df[column_names].replace(np.NaN, 'Unknown')

    val_list = ['Unknown']
    val_list = val_list + list(col_times.keys())

    le = LabelEncoder()
    le.fit(val_list)

    for c in column_names:
        df[c] = le.transform(df[c])

    df_part = df[column_names]


    colDict = df_part.to_dict('index')
    h = FeatureHasher(n_features=10)

    f = h.transform(colDict.values())

    hashNames = get_names((col+'_hash'), 10)
    df1 = pd.DataFrame(f.toarray(),
                       columns = hashNames)

    df = pd.concat([df, df1], axis=1)
    df = df.drop(column_names, axis=1)
df.head()
```
